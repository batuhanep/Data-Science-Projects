```python
# Importing necessary libraries
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Ok")
```

```python
#Python allows you to visualize data without the need for long codes and advanced coding knowledge. In this article, I tried to explain data visualization with Python from the perspective of a learner, with the notes I took while learning.
```

```python
#Reading csv file
sleep_data = pd.read_csv("/kaggle/input/sleep-health-and-lifestyle-dataset/Sleep_health_and_lifestyle_dataset.csv")
```

```python
#First five rows of data
#You can change the number of rows you want to display by entering numbers inside the Head. For example, head(6)
sleep_data.head()
```

```python
#SCATTER PLOTS
#Scatter Plot can be used to view negative or positive correlation between variables.

# Plot the data

# Width and height
# plt.figure helps to be able to adjust the size of the visualization.
#plt.figure(figsize=(10,6))

#Tittle
plt.title("Relationship Between QS and SL")

#Scatter Plot
sns.scatterplot(x=sleep_data['Stress Level'], y=sleep_data['Quality of Sleep'])
```

```python
# As we can see, Stress Level and Quality of Sleep negatively correlated.
#if stress level is higher, people also tend to sleep less.
```

```python
#Regression line is used to display the relationship between two variables more clearly.

#Regplot
sns.regplot(x=sleep_data['Stress Level'], y=sleep_data['Quality of Sleep'])

#Title
plt.title("Relationship Between QS and SL with Regression Line")

#
```

```python
#Scatter Plots can also be used to show the relationship between more than two variables. For example, we can add which sleep disorder relates to the relationship between sleep quality and stress level.

#hue: Hue is used to show the color distribution of the third added variable, sleep disorder.
sns.scatterplot(x=sleep_data['Stress Level'], y=sleep_data['Quality of Sleep'],hue=sleep_data['Sleep Disorder'] )
```

```python
#Let's use lmplot to display this relationship more clearly. 
#Sleep apnea has a steeper curve compared to insomnia.

sns.lmplot(x="Stress Level", y="Quality of Sleep", hue="Sleep Disorder", data=sleep_data)
```

```python
#Let's say we want to view the age distribution density of the sleep health dataset.
#We can use the sns.kdeplot method to display the density distribution.

#shade=True allows coloring the area below the line.

sns.kdeplot(sleep_data['Age'], shade=True)

#Title
plt.title("Age Distribution")
```

```python
#Differen dataset, Berlin Crimes
crime_data=pd.read_csv("/kaggle/input/crime-in-berlin-2012-2019/Berlin_crimes.csv", index_col="Year")
```

```python
crime_data.head()
```

```python
#BAR CHARTS

#For example, the distribution of the number of robberies by district can be examined with a bar chart.
sns.barplot(x=crime_data['District'], y=crime_data['Robbery'])

#Title
plt.title("Robbery rates by districts")

# xticks rotates the x-axis labels by 90 degrees, so that the texts do not overlap each other.
plt.xticks(rotation=90)

# Size
plt.figure(figsize=(10, 6))
```

```python
#Differen dataset, Diabetes
diabetes_data = pd.read_csv("/kaggle/input/diabetes-healthcare-comprehensive-dataset/health care diabetes.csv")
```

```python
diabetes_data.head()
```

```python
#HEATMAPS
#With the heatmap we can see how strong the correlation is between different variables.

#Correlation
diabetes_cor=diabetes_data.corr()

#Heatmap
sns.heatmap(diabetes_cor, annot=True)
#annot=true: It makes the value in each cell visible.

#Title
plt.title("Diabetes Heatmap")
```
