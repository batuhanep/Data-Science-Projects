```python
# G8 nations' collective share of the Global Domestic Product
# total gdp of g8 countries
g8_nations = ["France", "Germany", "Italy", "Japan", "United Kingdom", "United States", "Canada", "Russia"]
g8_total_gdp = country_pv[country_pv["country"].isin(g8_nations)]['GDP: Gross domestic product (million current US$)'].sum()
g8_total_gdp
```

```python
# total gdp
total_gdp = country_pv["GDP: Gross domestic product (million current US$)"].sum()
total_gdp
```

```python
#G8 nations' collective share of the Global Domestic Product(GDP)    
g8_gdp_share = (g8_total_gdp / total_gdp) * 100
g8_gdp_share
```

```python
print("G8 nations' collective share of the Global Domestic Product(GDP) is ", str(g8_gdp_share))
```

```python
# Population densities of each continent
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def map_to_continent(Region):
    if Region in ['SouthernAsia', 'SouthernEurope', 'WesternAsia', 'EasternAsia', 'CentralAsia']:
        return 'Asia'
    elif Region in ['NorthernAfrica', 'MiddleAfrica', 'WesternAfrica', 'EasternAfrica', 'SouthernAfrica']:
        return 'Africa'
    elif Region in ['SouthAmerica', 'CentralAmerica', 'Caribbean']:
        return 'America'
    elif Region in ['NorthernAmerica']:
        return 'North America'
    elif Region in ['WesternEurope', 'EasternEurope', 'NorthernEurope']:
        return 'Europe'
    elif Region in ['Polynesia', 'Melanesia', 'Micronesia']:
        return 'Oceania'
    else:
        return 'Other'

country_pv['Continent'] = country_pv['Region'].apply(map_to_continent)
```

```python
country_pv
```

```python
sns.boxplot(data=country_pv, x="Continent", y="Population density (per km2, 2017)")
plt.xlabel("Continent")
plt.ylabel("Population Density per km2")
plt.title("Population Density by Continent")
plt.xticks(rotation=60)
plt.show()
```

```python
# List the first ten countries with the maximum governmentexpenditure on education (% of GDP).
sorted_education = country_pv.sort_values(by="Education: Government expenditure (% of GDP)", ascending=False)
top_ten_education = sorted_education.head(10)
print(top_ten_education[['country', 'Education: Government expenditure (% of GDP)']]) 
```

```python
# countries have population more than %50
urban_over_50 = country_pv[country_pv['Urban population (% of total population)'] > 50]
urban_over_50
```

```python
urban_pop_50=country_pv[country_pv["Urban population (% of total population)"]>50]

western_eu=urban_pop_50=urban_pop_50[urban_pop_50["Region"]=="WesternEurope"]

western_eu=western_eu.sort_values(by="Urban population growth rate (average annual %)", ascending=False)

for index, row in western_eu.iterrows():
    country=row["country"]
    growth=row["Urban population growth rate (average annual %)"]
    
    years=range(1,11)
    urban_pop=[row["Urban population (% of total population)"] * (1 + float(growth)/100) ** (year-1) for year in years]

    plt.plot(years, urban_pop, label=country)
    
plt.title("urban population trend curves in the next ten years for the Western European countries")
plt.xlabel("year")
plt.ylabel("urban population(%)")
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
# Compare the Area, Population, and Population Density information with the information retrieved from the Wikipedia Caribbean page (get this information through Wikipedia) for the Caribbean countries.

import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Caribbean'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
caribbean_titles = table.find_all('th')
```

```python
caribbean_titles
```

```python
caribbean_table_titles=[title.text.strip() for title in caribbean_titles]

print(caribbean_table_titles)
```

```python
df = pd.DataFrame(columns = caribbean_table_titles)

df
```

```python
# drop total column

df.drop(columns=['Total'], inplace=True)
```

```python
df
```

```python
column_data=table.find_all('tr')

column_data
```

```python
for i, row in enumerate(column_data[1:], 1):
    row_data = row.find_all('td')
    if len(row_data) == len(df.columns):
        individual_row_data = [data.text.strip() for data in row_data]
        df.loc[len(df)] = individual_row_data
```
