```python
#Let's continue with openpowerlifting dataset
powerlifting = pd.read_csv("openpowerlifting.csv")
powerlifting
```

```python
# First five columns
powerlifting.head()
```

```python
# Maps
# Suppose that we want to remean BodyweightKg column values by using map

powerlifting_bodyweightkg_mean = powerlifting.BodyweightKg.mean()
powerlifting.BodyweightKg.map(lambda p: p-powerlifting_bodyweightkg_mean)

# In this code we subtracted the average value from each body weight. Thus, we can see how much an athlete's body weight is below or above average.
#In Python Pandas, lambda is a function used to operate on a DataFrame. Lambda makes functions more concise.
```

```python
# Maps with def and apply

def remean_bodyweight(kg):
    kg.BodyweightKg = kg.BodyweightKg - powerlifting_bodyweightkg_mean
    return kg
powerlifting.apply(remean_bodyweight, axis='columns')
#We can apply the function we created with Apply( , ) to the dataset.
```

```python
#Faster way for remeaning bodyweight values
powerlifting_bodyweightkg_mean = powerlifting.BodyweightKg.mean()
powerlifting.BodyweightKg - powerlifting_bodyweightkg_mean
```

```python
#Combining columns
powerlifting.Name + " - " + powerlifting.Sex
```

```python
# GROUPING AND SORTING

#Group by WeightClassKg
#Just like in SQL :)
powerlifting.groupby('WeightClassKg').WeightClassKg.count()
```

```python
# Let's run different functions together by using groupby
powerlifting.groupby(['BodyweightKg']).Age.agg([len, min, max,])
```

```python
#Multi-indexes
powerlifting_multi=powerlifting.groupby(['BodyweightKg','Division']).BestDeadliftKg.agg([max])
powerlifting_multi
```

```python
# Different way to do this with lambda and idxmax
powerlifting = powerlifting.dropna()
powerlifting.groupby(['BodyweightKg']).apply(lambda df: df.loc[df.BestDeadliftKg.idxmax()])
#The idxmax() method returns a Series with the index of the maximum value for each column.
```

```python
# Reseting multi-indexes

powerlifting_multi.reset_index()
```

```python
#Sorting data
#If we want to rank with the best deadlift record
powerlifting_multi= powerlifting_multi.reset_index()
powerlifting_multi.sort_values(by='max')
```

```python
#Descending order by adding 'ascending=False'
powerlifting_multi.sort_values(by='max', ascending=False)
```

```python
#Joining, Combining Datasets
cnd_youtube = pd.read_csv("CAvideos.csv")
brit_youtube = pd.read_csv("GBvideos.csv")

pd.concat([cnd_youtube, brit_youtube])
```

```python
#Let's join the data sets like in SQL. Pandas allows us to join two datasets with short codes.
left = cnd_youtube.set_index(['title', 'trending_date'])
right = brit_youtube.set_index(['title', 'trending_date'])

right.join(left, lsuffix='_CAN', rsuffix='_UK')
```
